### **Ejercicio 2**

```python
def calculos_mat(calif):
    calif = pd.Series(calif)
    calculos = pd.Series([calif.min(), calif.max(), calif.mean(), calif.std()], index = ['Min', 'Max', 'Media', 'Desviación típica'])
    # calif.min, max, mean, std, etc son funciones basicas de la libreria pandas para el calculo de esos valores
    return calculos

calif = {'Erick':55, 'Gio':45, 'Jesus':36, 'Juan': 100, 'Joel': 98, 'Wilbert':88, 'Fernando':99 , 'Santiago':10}
print(calculos_mat(calif))
```

### **Ejercicio 8** 

##### 1. Generar un DataFrame con los datos de los cuatro ficheros.

```python
import numpy as np #invocamos las librerias
import datetime as dt

emisiones_2016 = pd.read_csv('https://raw.githubusercontent.com/ErickMatoss/archivospruebas/main/emisiones-2016.csv', sep = ';')
emisiones_2017 = pd.read_csv('https://raw.githubusercontent.com/ErickMatoss/archivospruebas/main/emisiones-2017.csv', sep = ';')
emisiones_2018 = pd.read_csv('https://raw.githubusercontent.com/ErickMatoss/archivospruebas/main/emisiones-2018.csv', sep = ';')
emisiones_2019 = pd.read_csv('https://raw.githubusercontent.com/ErickMatoss/archivospruebas/main/emisiones-2019.csv', sep = ';')
emisiones = pd.concat([emisiones_2016, emisiones_2017, emisiones_2018, emisiones_2019]) #concatenamos los 4 dataframes en 1
emisiones
```

##### 2. Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, etc.

```python
column = ['ESTACION', 'MAGNITUD', 'ANO', 'MES']
column.extend([col for col in emisiones if col.startswith('D')])
emisiones = emisiones[column]
emisiones
```

#### 3. Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.

```python
emisiones = emisiones.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='VALOR')
emisiones
```

##### 4.  Crear una nueva columna con las fechas a partir del año, mes y día

```python
emisiones['DIA'] = emisiones.DIA.str.strip('D') #se elimina la D de dia
#se concatenan las columnas del año, mes y día
emisiones['FECHA'] = emisiones.ANO.apply(str) + '/' + emisiones.MES.apply(str) + '/' + emisiones.DIA.apply(str)
#actualizamos la nueva columna al con typo fecha
emisiones['FECHA'] = pd.to_datetime(emisiones.FECHA, format='%Y/%m/%d', infer_datetime_format=True, errors='coerce')
emisiones
```

#### 5. Eliminar las filas con fechas no válidas (utilizar la función isnat del módulo numpy) y ordenar el DataFrame por estaciones, contaminantes y fecha

```python
emisiones = emisiones.drop(emisiones[np.isnat(emisiones.FECHA)].index) #eliminamos las fechas invalidas
emisiones.sort_values(['ESTACION', 'MAGNITUD', 'FECHA']) # con este comando ordenamos el df en base a los valores dados
```

##### 7. Crear una función que reciba una estación, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estación y rango de fechas dado.

```python
def evolucion(estacion, contaminante, desde, hasta):
    evol = emisiones[(emisiones.ESTACION == estacion) & (emisiones.MAGNITUD == contaminante) & (emisiones.FECHA >= desde) & (emisiones.FECHA <= hasta)].sort_values('FECHA').VALOR
    return evol
print(evolucion(56, 8, dt.datetime.strptime('2017/08/16', '%Y/%m/%d'), dt.datetime.strptime('2019/07/24', '%Y/%m/%d')))
```

##### 10. Crear una función que reciba una estación y un contaminante y devuelva un resumen descriptivo de las emisiones del contaminante indicado en la estadión indicada.

```python
def resumen(estacion, contaminante):
  emi = emisiones[(emisiones.ESTACION == estacion) & (emisiones.MAGNITUD == contaminante)].VALOR.describe()
  return emi 

print('Resumen Dióxido de Nitrógeno en la estacion 60:\n', resumen(60, 8),'\n', sep='')

print('Resumen Dióxido de Nitrógeno en la estacion 35:\n', resumen(35, 8), sep='')
```

#### 11. Crear una función que devuelva las emisiones medias mensuales de un contaminante y un año dados para todos las estaciones.


```python
def evolucion_mensual(contaminante, año):
  evolucion = emisiones[(emisiones.MAGNITUD == contaminante) & (emisiones.ANO == año)].groupby(['ESTACION', 'MES']).VALOR.mean().unstack('MES')
  return evolucion

# Evolución del dióxido de nitrógeno en 2018
evolucion_mensual(8, 2018)
```

```python
def med_mensual(est):
  med_men = emisiones[(emisiones.ESTACION == est)].groupby(['ESTACION', 'MAGNITUD']).VALOR.mean().unstack('MAGNITUD')
  print("Media con istintos tipos de contaminantes en la estacion", est)
  return med_men

esta = int(input("Ingresa la estacion: "))
med_mensual(esta)
```
